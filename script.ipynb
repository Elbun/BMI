```python
## BMI
# Import Libraries
import numpy as np
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
```

```python
# Get the train data
train_dataset = pd.read_csv("bmi_data/bmi_train.csv")
train_dataset
```

```python
# Get the summary of train data (1)
print("dataset size (rows, columns) =", train_dataset.shape)
train_dataset.info()
```

```python
# Get the summary of train data (2)
train_dataset["Gender"].unique()
```

```python
# EDA (1) - count gender
train_dataset.groupby("Gender").count()
```

```python
# EDA (2) - distribution of weight
# Generate the density plot 
chart = alt.Chart(train_dataset).transform_density( 
        density='Weight', 
        as_=['Weight', 'density'], 
        groupby=['Gender'],   
        extent=[0, 220] 
    ).mark_area(
        opacity=0.5
    ).encode(    
        x=alt.X('Weight:Q', title='Weight'), 
        y=alt.Y('density:Q', title='Density'), 
        color=alt.Color('Gender:N', scale=alt.Scale(domain=train_dataset['Gender'].unique(), range=['blue', 'red']), title='Gender'),
        tooltip=['Gender:N', 'Weight:Q', 'density:Q']
    ).properties(title='Age Density Plot of Titanic Travellers', width=300, height=200
    ).facet('Gender:N', columns=2
    ) 

# Render the chart
chart.display()
```

```python
# EDA (2) - distribution of height
# Generate the density plot 
chart = alt.Chart(train_dataset).transform_density( 
        density='Height', 
        as_=['Height', 'density'], 
        groupby=['Gender'],
        extent=[100, 220] 
    ).mark_area(
        opacity=0.5
    ).encode(    
        x=alt.X('Height:Q', title='Height'), 
        y=alt.Y('density:Q', title='Density'), 
        color=alt.Color('Gender:N', scale=alt.Scale(domain=train_dataset['Gender'].unique(), range=['blue', 'red']), title='Gender'),
        tooltip=['Gender:N', 'Height:Q', 'density:Q']
    ).properties(title='Age Density Plot of Titanic Travellers', width=300, height=200
    ).facet('Gender:N', columns=2
    ) 

# Render the chart
chart.display()
```

```python
# EDA (3) - Index by Weight and Height
# Generate the heatmap
chart = alt.Chart(train_dataset).mark_rect().encode(
        x = alt.X('Weight:O'),
        y = alt.Y('Height:O'),
        color = 'Index:Q'
    ).properties(title='Age Density Plot of Titanic Travellers', width=600, height=400
    )

# Render the chart
chart.display()
```

```python
# EDA (3) - Index by BMI
# Generate the scatterplot
train_dataset["BMI"] = train_dataset["Weight"] / ((train_dataset["Height"]/100)*(train_dataset["Height"]/100))
train_dataset

## Chart 15a
chart = alt.Chart(train_dataset[train_dataset["Gender"]=="Male"]).mark_point(size=20).encode(
    x=alt.X("BMI:Q", title="BMI", axis=alt.Axis(labelAngle=-45)).scale(zero=False),
    y=alt.Y("Index:Q", title="Index").scale(zero=False),
    color = 'Gender:N'
)
final_plot = chart + chart.transform_regression("BMI","Index").mark_line(color="red")
fig = (final_plot).configure_axis(
            labelFontSize=10
        ).properties(
            title=('Avocado Sales Volume vs Price (Conventional Type)'),
            width=400,
            height=400
        )

# Render the chart
fig.display()
```
